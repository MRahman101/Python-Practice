```python
# Arithmetic:+, -, *, / (returns a float), // (int division), % (modulus aka remainder)
friends=10-5
print(friends)
#augmented operator: +=, /=, etc.
friends += 3
print(friends)
friends /= 3
print(friends)
```

```python
#If, Elif, Else
age=input("What is your age? ")
age=int(age)
if age >= 18:
  print("You are an adult")
elif age >= 13:
  print("You are a teenager")
else:
  print("You are a child")
```

```python
#Logical Operators (or, and, not)
age = 18
has_ticker = False
is_vip = True
# Must be 18+ AND (have a ticket OR be VIP) AND the last condition must not be False
if age >=18 and (has_ticker or is_vip) and not False:
  print("You can enter")
else:
  print("You cannot enter")
```

```python
#While loop
name = input("Enter your name: ")
while name == "":
  name = input("Enter your name: ")
age = int(input("Enter your age: "))
while age == 0:
  age = int(input("Enter your age again: "))

print(f"Hello {name}!")
```

```python
#For loops
for i in range (4): #For i (index) in range (this is a function)
  print(i) #this prints the index/counter for repeats and start at index value 0.
for i in range (50,60,2): #Range of index from index 50 to 60 & increment by 2.
  print(i)
name= "John"
for letter in name:
  print(letter, end=" ") #end makes the letters arrange in horzontal not vertical (default)
print("\n")
for i in range (5,1,-1):#loop backwards. start index 5 and loop backward to i 1.
  print(i)
```

```python
#Python Data Structures
#List: [] ; "Packing Boxes"; Mutable, Ordered, Duplicate
#Tuple: () ; "Handcuffs" ; Immutable, Ordered, Duplicate, (faster than list)
#Dictionary: {k:v} ; "Keys to kingdom" ; Mutable, Ordered, Duplicate (Values only)
#Set: {} ; "VIP Club" ; Mutable, Unordered ; No Duplicate allowed
fruits = ["apple", "banana", "cherry"]
print(fruits[1])
fruits[1] = "grapes"
for fruit in fruits:
  print(fruit)
fruits.append("orange") #add element to list
print(fruits)
fruits.pop(0) #remove element from list
print(fruits)

info = {name: "john", age: 23, is_student: True}
print(info[name])
```
